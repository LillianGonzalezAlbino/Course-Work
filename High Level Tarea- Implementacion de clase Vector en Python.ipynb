### Ejercicio 1:  
Queremos desarrollar una clase de Python3 que permita manejar arreglos unidimensionales de números reales. Para eso, tenemos la especificación de las clases a continuación. Implemente los métodos propuestos. Escriba un programa sencillo que permita verificar el correcto funcionamiento de la clase Vector.

```python
class SizeMimatch (Exception):
    pass
 
class IllegalSize (Exception):
    pass
 
class IllegalIndex(Exception):
    pass

#Clase que maneja vectores (arreglos unidimensionales) de numeros floating-point
class Vector:

    #Constructor que recibe la cantidad de elementos. Verifica si un tamano 
    #ilegal es dado. Es decir, si el tamano es menor que 1, la excepcion 
    #Illegal sera alzada o "raised". Esta funcion no devuelve nada
    def __init__(self, size):
        
        #verificar si el tamano dado es menor que 1
        if(size < 1):
            
            #si es menor que 1, alzar la excepcion IllegalSize
            raise IllegalSize ('Incorrect size')
        
        #si el valor es valido, crear vector del tamano size compuesto de None
        else:
            
            #si el valor dado en size no es un int, hacer un 
            #casting de size a int
            self.size = int(size)
            self.vector = [None]*size
    
    #Metodo que devuelve el valor de size (tamano del vector)
    #Esta funcion devuelve el tamano del vector
    def getSize(self):
        
        return self.size
    
    #Metodo que devuelve el elemento en la posicion i del vector. 
    #Se verifica si el valor de i es un valor legal, si no lo es, 
    #se alza la excepcion IllegalIndex
    #Para que sea un valor legal debe ser positivo y menor que size.
    #Esta funcion devuelve un elemento del vector
    def get(self, i):
        
        #si el valor de i es negativo o mayor o igual a size, alza 
        #excepcion IllegalIndex
        if(i < 0 or i >= self.size):
            raise IllegalIndex
        
        #si i no es un valor ilegal, entonces devolver el elemento 
        #en la posicion i del vector
        else:
            
            return self.vector[i]
        
    #Metodo que pone el valor value en la posicion i del vector. 
    #Se verifica si i es un valor legal, es decir si es positivo y menor que size. 
    #De no ser un valor legal se alza la excepcion IllegalIndex
    #Esta funcion no devuelve nada
    def set(self, i, value):
        
        #si el valor de i es negativo o mayor o igual a size, alza 
        #excepcion IllegalIndex
        if(i < 0 or i >= self.size):
            raise IllegalIndex ('Wrong index')
            
        #si i no es un valor ilegal, entonces agrega el valor value en 
        #la posicion i del vector
        else:
            
            self.vector[i] = value
            
    #Este metodo sobrecarga el operador + y anade dos vectores elemento 
    #por elemento. Es decir, lo elementos que tengan el mismo indice, 
    #seran anadidos. Si los tamanos de los vectores no son iguales, la 
    #excepcion SizeMismatch sera alzada. Esta funcion devuelve un vector
    def __add__(self, other):
        
        #Verificar que los tamanos de los vectores sean iguales
        if (self.size == other.size):
            
            #crear vector vec donde se guardara la suma de self y other
            vec = Vector(self.size)
            
            #ir elemento por elemento en cada vector para sumarlos
            for i in range(0, self.size):
                
                vec.vector[i] = self.vector[i] + other.vector[i]
                
            #devolver suma de vectores
            return vec
        
        #Si los tamanos no son iguales, se alza la excepcion SizeMismatch
        else:
            
            raise SizeMismatch ('To add vectors they must be the same size')
        
    #Este metodo sobrecarga el operador - y resta dos vectores elemento 
    #por elemento. Es decir, lo elementos que tengan el mismo indice, 
    #seran restados. Si los tamanos de los vectores no son iguales, la excepcion 
    #SizeMismatch sera alzada. Esta funcion devuelve un vector
    def __sub__ (self, other):

        #Verificar que los tamanos de los vectores sean iguales
        if(self.size == other.size):
            
            #crear vector vec donde se guardara la resta de self y other
            vec = Vector(self.size)
            
            #ir elemento por elemento en cada vector para restarlos
            for i in range(0, self.size):
                
                vec.vector[i] = self.vector[i] - other.vector[i]
            
            #devolver suma de vectores
            return vec
        
        #Si los tamanos no son iguales, se alza la excepcion SizeMismatch
        else:
            
            raise SizeMismatch ('To subtract vectors they must be the same size')
            
    #Este metodo sobrecarga el operador * y verifica si el argumento other es un
    #vector o si es un escalar. Si es un vector, multiplica los dos vectores elemento 
    #por elemento. Es decir, lo elementos que tengan el mismo indice, seran 
    #multiplicados. Si los tamanos de los vectores no son iguales, la excepcion 
    #SizeMismatch sera alzada. Esto devuelve un vector
    #Si other es un escalar, devuelve un vector nuevo que consiste de
    #el vector viejo multiplicado por un escalar
    def __mul__ (self, other):

        #multiplicacion por un escalar
        #Ejercico 4 de la asignacion se incluyo en la definicion del metodo __mul__ 
        if(isinstance(other, int) or isinstance(other, float)): 

            #crear variable donde se guardara la copia del vector
            vec = Vector(self.size)
        
            #ir elemento por elemento del vector asigandolos al vector nuevo y 
            #multiplicandolo por el escalar
            for i in range (0, self.size):
            
                vec[i] = other * self.vector[i]
        
            #devolvemos el vector
            return vec
        
        #multiplicacion por un vector
        else:
        
            #Verificar que los tamanos de los vectores sean iguales
            if(self.size == other.size):
            
                #crear vector vec donde se guardara la multiplicacion de self y other
                vec = Vector(self.size)
            
                #ir elemento por elemento en cada vector para multiplicarlos
                for i in range(0, self.size):
                
                    vec.vector[i] = self.vector[i] * other.vector[i]
                
                #devolver multilplicacion de vectores
                return vec
        
            #Si los tamanos no son iguales, se alza la excepcion SizeMismatch
            else:
            
                raise SizeMismatch ('To multiply vectors they must be the same size')
        
            
    #Este metodo sobrecarga el operador == y compara dos vectores elemento 
    #por elemento. Es decir, lo elementos que tengan el mismo indice, seran 
    #comparados. Si los tamanos de los vectores no son iguales, la excepcion 
    #SizeMismatch sera alzada. Esta funcion devuelve valores booleanos 
    #(True or False)
    def __eq__ (self, other):

        #Verificar que los tamanos de los vectores sean iguales
        if(self.size == other.size):
            
            #ir elemento por elemento en cada vector para verificar si son iguales
            for i in range(0, self.size):
                
                #si los elementos de los vectores no son iguales, devuelve False
                if(self.vector[i] != other.vector[i]):
                    return False
                
            #Si termina el for loop, entonces todos los elementos son iguales y 
            #devolvemos True
            return True
        
        #Si los tamanos no son iguales, se alza la excepcion SizeMismatch
        else:
    
            raise SizeMismatch ('To check if vectors are equal they must be the same size')
      
    #Este metodo sobrecarga el operador != y compara dos vectores elemento 
    #por elemento. Es decir, lo elementos que tengan el mismo indice, seran 
    #comparados. Si los tamanos de los vectores no son iguales, la excepcion 
    #SizeMismatch sera alzada. Esta funcion devuelve valores booleanos 
    def __ne__ (self, other):
        
        #Note que al devolver lo opuesto de la comparacion de self y other (si 
        #son iguales) ya se cumplen las condiciones establecidas ya que la 
        #funcion definida para sobrecargar el operador == ya verifica si los 
        #tamanos de los vectores son iguales y si no lo son alza la excepcion 
        #SizeMismatch
        if(self == other):
            
            return False
        
        else:
            
            return True

    
    #Este metodo sobrecarga el operador unario + que devuelve una copia del 
    #vector. Se crea otro espacio en memoria de tal manera que ambos vectores 
    #no esten apuntando al mismo vector en memoria. De esta manera se obtienen 
    #dos vectores similares. 
    def __pos__ (self):

        #crear variable donde se guardara la copia del vector
        vec = Vector(self.size)
        
        #ir elemento por elemento del vector asigandolos al vector nuevo
        for i in range (0, self.size):
            
            vec[i] = self.vector[i]
        
        #devolvemos la compia del vector
        return vec
        
    
    #Este metodo sobrecarga el operador unario + que devuelve una copia del 
    #vector con cada elemento multiplicado por -1. 
    #Se crea otro espacio en memoria de tal manera que ambos vectores no 
    #esten apuntando al mismo vector en memoria. De esta manera se obtienen 
    #dos vectores.
    def __neg__(self):

        #crear variable donde se guardara la copia del vector
        vec = Vector(self.size)
        
        #ir elemento por elemento del vector asigandolos al vector nuevo y 
        #multiplicandolo por -1
        for i in range (0, self.size):
            
            vec[i] = -self.vector[i]
        
        #devolvemos el vector 
        return vec

    
    #Este metodo devuelve el valor minimo dentro del vector
    #Se utiliza un for loop para comparar cada elemento
    #Esta funcion devuleve un elemento del vector
    def min (self):
        
        #establecer el primer elemento como el valor minimo
        min = self.vector[0]
        
        #ir elemento por elemento verificando si alguno es menor que el minimo 
        #establecido
        for i in range (0, self.size):
            
            #si el valor en la posicion i es menor que el minimo se reasigna el 
            #valor minimo a el valor en la posicion i
            if(self.vector[i] < min):
                
                min = self.vector[i]
                
        #se devuleve el valor minimo        
        return min 
    
    #Este metodo devuelve el valor maximo dentro del vector
    #Se utiliza un for loop para comparar cada elemento
    #Esta funcion devuleve un elemento del vector
    def max (self):
        
        #establecer el primer elemento como el valor maximo
        max = self.vector[0]
        
        #ir elemento por elemento verificando si alguno es mayor que el maximo 
        #establecido
        for i in range (0, self.size):
            
            #si el valor en la posicion i es mayor que el maximo se reasigna el 
            #valor maximo a el valor en la posicion i
            if(self.vector[i] > max):
                
                max = self.vector[i]
                
        #se devuleve el valor maximo        
        return max
     
    #Metodo que devuleve la suma de todos los elementos en el vector
    #Esta funcion devuelve un numero real
    def sum (self):
        
        #establecer variable que guarda resultado
        resultado = 0.0
        
        #ir elemento por elemento sumandoselo a la variable resultado
        for i in range (0, self.size):
            
            resultado += self.vector[i]
            
        #devolver el resiltado
        return resultado
    
    #Metodo que devulece el promedio de los elementos del vector
    def average (self):

        #establecer variable que guarda resultado de la suma de todos los elementos 
        #utilizando metodo sum
        resultado = self.sum
        
        #devolvemos el resultado dividido entre el tamano que nos da el promedio
        return (resultado/self.size)
```

Agregue los métodos que sean necesarios a la clase Vector para sobrecargar su representación canónica y su representación bonita.

```python
    #Este metodo es la representcion canonica para todos los objetos de vectores.
    #Cuando el objeto es llamado se devolvera el nombre de la clase con el tamano 
    def __repr__ (self):
        
        #se devuelve el nombre de la clase con el tamano i.e. Vector([size])
        return "Vector({})".format(self.size)
    
    #Este metodo es la representcion bonita para todos los objetos de vectores.
    #Se devuelve un string con el nombre de la clase y el listado de todos los 
    #elementos del vector
    #i.e. Vector[] = { value_1, value_2, ... , value_size }
    def __str__ (self):
        
        #creamos un string con el principio de nuestra representacion bonita
        string = "Vector[]={"
        
        #hacemos un for loop donde agregaremos los elementos del vector al string 
        #sin incluir el ultimo elemento
        for i in range (0, self.size-1):
            string += "{}, ".format(self.vector[i])
            
        #ahora agregamos el ultimo elemento y la llave que cierra nuestro string 
        string += "{}".format(self.vector[self.size-1]) + "}"
        
        #devolvemos el string de nuestra representacion bonita
        return string
```

### Ejercicio 5:
Agregue los método que sean necesarios a la clase Vector para que se puedan utilizar los operadores +=, -= y *=. Estos operadores son también conocidos como operadores “in-place”.

```python
    #Este metodo sobregarga el operador += 
    #Note que al usar + ya verifica si el tamano de los vectores es igual
    #y si no lo son alza la excepcion SizeMismatch
    #note que al operador + ya estar sobrecaragdo, es suficiente 
    #regresar la suma de los vectores self y other
    def __iadd__(self, other):
        return self + other
    
    #Este metodo sobregarga el operador -= 
    #Note que al usar - ya verifica si el tamano de los vectores es igual
    #y si no lo son alza la excepcion SizeMismatch
    #note que al operador - ya estar sobrecaragdo, es suficiente 
    #regresar la resta de los vectores self y other
    def __isub__(self, other):
        return self - other
    
    #Este metodo sobregarga el operador *= 
    #Note que al usar * ya verifica si el tamano de los vectores es igual
    #y si no lo son alza la excepcion SizeMismatch
    #note que al operador * ya estar sobrecaragdo, es suficiente 
    #regresar la multiplicacion de los vectores self y other
    def __imul__(self, other):
        return self * other
```
